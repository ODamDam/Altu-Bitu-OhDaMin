```python
bgr_img = cv2.imread('./porygon.jpg')
print(bgr_img.shape) # y:600, x: 600, c:3
print(bgr_img.dtype) #unit8 = unsigne char = 1byte
print(bgr_img.size) # = 600*500*3=1.08MB
```

```python
import torch
```

```python
plt.imshow(cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB))
```

```python
#사이즈 반토막 내보기
#numpy만 이용해서
half_img = bgr_img[::2, ::2]
print(half_img.shape)

#정답
#half_img = cv2.resize(bgr_img, (0,0), fx=0.5, fy=0.5)
plt.imshow(cv2.cvtColor(half_img, cv2.COLOR_BGR2RGB))
plt.show
```

```python
img_gray = 0.2126 * bgr_img[:, :, 2] \
  + 0.7152 * bgr_img[:, :, 1] + 0.0722 * bgr_img[:, :, 0]
img_gray=img_gray.astype(np.uint8)
print(img_gray.shape)
plt.imshow(img_gray, cmap='gray')
```
